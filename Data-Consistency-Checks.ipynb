```python
# so the MTL_EC05-LAB showed up on this list too, and we will keep it. otherwise there appears to be some formatting inconsistency here...
# going to take a closer look at stations with 75 or less entries
# as end_stations is a string, and station_id is a code with string len =5,  going to find all entries with a string len > 6


end_station_df = df1[df1['end_station_id'].apply(len) >= 6]
end_station_df.shape
```

```python
# end station_df appears to have 996 incorrectly formatted entries:
# going to exclude these from our working df, however the MTL-EC0-LAB is consistent across start & end, so we will keep this.
# forgot to resent the display, don't need this many cells.

pd.set_option('display.max_rows', 60)

end_station_df
```

```python
# going to use .apply and lambda method to keep the entries we want:
# MTL-EC05-LAB is a string length of 12, so we want the string length to be less than 5 (to match the rest), but at least 10.

# explaining what the lambda is doing:
# it is saying that the length of x must be in our window. X is the specific entry in our end station id column. 
# searching for the # of entries we want to make sure stay in our new DF

df1[df1['end_station_id'].apply(lambda x: (5 <= len(x) >= 10))]
```

```python
# these are the 7 entries we want to make sure to keep
# (hopefully we can find Shop Morgan in start_station and assign the end station_id

# our df1 has a size of 892281 x 15, and the end_station_df is 996 x 15. We know we want to keep 7 entries from that, so we willexpect our new df to be 891292 x 15
```

```python
# using apply and lambda again
# this time going to make sure we only KEEP entries with len 5 or less OR len 10 or more

df2 = df1[df1['end_station_id'].apply(lambda x: len(x) < 6 or len(x) > 10)]
df2
```

```python
# now that we've dropped incorrectly formatted end_stations, going to do a quick check for 'Shop Morgan'

df2[df2['start_station_name'] == 'Shop Morgan']
```
